```python
import pandas as pd

df = pd.DataFrame(columns=['Name', 'Type'])
df.head()

def generate_hash_key(var_name):
    ascii_value = 0

    for i in var_name:
        ascii_value += ord(i)
    return ascii_value % 100

def insert(df, name, typ):
    if (name not in df.Name.values) or (typ not in df.Type.values):
      df.loc[generate_hash_key(name)]=[name, typ]
      print('Successfully Inserted.')

    else:
      print('Double Input! Error.')

    return df

def search(df,name):
    key = generate_hash_key(name)
    return df.loc[key]

def delete(df, name):
    try:
        key = generate_hash_key(name)
        df = df.drop(key)
        print('Successfully Deleted')
        return df

    except:
        print('Name not found')
        return df
    
def update(df, name, new_name, typ, new_typ):
    if name:
        if name in df.Name.values:
            df.loc[generate_hash_key(name)].Name = new_name
    if typ:
        if typ in df.Type.values:
            df.loc[generate_hash_key(name)].Type = new_typ
            
    return df

def show():
    print(df)

while True:

  print("1. Insert")
  print("2. Search")
  print("3. Update")
  print("4. Delete")
  print("5. Show All")
  print("6. GetHashKey")
  print("7. Exit")
  
  i = int(input("Enter Choice: "))

  if i == 1:
    name = input("Enter Name: ")
    value = input("Enter Value: ")
    insert(df, name, value)

  elif i == 2:
    name = input("Enter Name: ")
    search(df, name)
    print(df)

  elif i == 3:
    old_name = input("Enter Old Name: ")
    new_name = input("Enter New Name: ")
    old_type = input("Enter Old Type: ")
    new_type = input("Enter New Type: ")
    df = update(df,old_name,new_name,old_type,new_type)
    print(df)
  
  elif i == 4:
    name = input("Enter Name: ") 
    df = delete(df, name)
  
  elif i == 5:
    print(df)

  elif i == 6:
    name = input("Enter Name: ") 
    print(generate_hash_key(name))
    
  elif i == 7:
    print("Process Ending")
    break  
```
